## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
male = count_gender["Male"]
male

female = count_gender["Female"]
female

nan = count_gender["Other / Non-Disclosed"]
nan

male_percentage = count_gender["Male"]/total_purchases*100
male_percentage

female_percentage = count_gender["Female"]/total_purchases*100
female_percentage

nan_percentage = count_gender["Other / Non-Disclosed"]/total_purchases*100
nan_percentage
        
genders_df = pd.DataFrame({"Concepts":["Male Players", "Female Players", "Non-Disclosed Players"],
                          "Count of":[male, female, nan],
                          "Percentages":[male_percentage,female_percentage,nan_percentage]})

genders_df["Percentages"] = genders_df["Percentages"].map("{:.2f}%".format)
genders_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
avg_pu = purchase_data[['SN','Price','Gender']]
avg_pu
average_pd = avg_pu.groupby("Gender", as_index=False).agg({"Price":["mean","sum"],"SN":"count"})
average_pd.columns

ex3 = avg_pu.drop_duplicates(subset=["SN"])
ex3 = ex3.groupby("Gender", as_index=False).agg({"SN":"count","Price":"sum"})
ex3
average_pd["Unique Count"]= ex3["SN"]
average_pd["Average Purchase Per User"] = average_pd[("Price", "sum")]/average_pd["Unique Count"]

average_pd[("Price", "mean")] = average_pd[("Price", "mean")].map("${:.2f}".format)
average_pd[("Price", "sum")] = average_pd[("Price", "sum")].map("${:.2f}".format)
average_pd["Average Purchase Per User"] = average_pd["Average Purchase Per User"].map("${:.2f}".format)

average_pd.set_index("Gender")

average_pd
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,50]

bins

group_names = ["<10", "10-14","15-19","20-24","25-29","30-34","35-39",">40"]

ex4_df = purchase_data[["Age", "SN", "Price"]]
ex4_df = ex4_df.drop_duplicates(subset=["SN"])
ex4_df["binning_summary"] = pd.cut(ex4_df["Age"], bins, labels=group_names, include_lowest=True)
ex4_df = ex4_df.groupby("binning_summary").count()
ex4_df["% of players"] = ex4_df['Age']/ex4_df['Age'].sum()*100
ex4_df

ex4_df = ex4_df[['Age','% of players']]

ex4_df['% of players'] = ex4_df['% of players'].map('{:.2f}%'.format)
ex4_df = ex4_df.rename(columns={'Age':'Total Count'})

ex4_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,50]

bins

group_names = ["<10", "10-14","15-19","20-24","25-29","30-34","35-39",">40"]

ex5_df = purchase_data[["Age", "SN", "Price"]]
ex5_df["binning_summary"] = pd.cut(ex5_df["Age"], bins, labels=group_names, include_lowest=True)
ex5_df2 = ex5_df.groupby("binning_summary").agg({"Age":"count", "Price":["mean","sum"]})
ex5_df = ex5_df.drop_duplicates(subset="SN").groupby("binning_summary").count()
ex5_df2["Average Total Purchase Per Person"] = ex5_df2[("Price","sum")]/ex5_df["SN"]
ex5_df2[("Price", "mean")] = ex5_df2[("Price", "mean")].map("${:.2f}".format)
ex5_df2[("Price", "sum")] = ex5_df2[("Price", "sum")].map("${:.2f}".format)
ex5_df2["Average Total Purchase Per Person"] = ex5_df2["Average Total Purchase Per Person"].map("${:.2f}".format)
ex5_df2
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_sn = purchase_data[["SN","Age","Price"]]
top_sn = top_sn.groupby("SN").agg({"Age":"count","Price":["mean", "sum"]})
top_sn = top_sn.sort_values(("Price","sum"), ascending=False)


top_sn[("Price", "mean")] = top_sn[("Price", "mean")].map("${:.2f}".format)
top_sn[("Price", "sum")] = top_sn[("Price", "sum")].map("${:.2f}".format)

top_sn = top_sn.head(5)
top_sn

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pop_item = purchase_data[["Item ID","Age","Item Name","Price"]]
pop_item = pop_item.groupby(["Item ID","Item Name"]).agg({"Age":"count","Price":["mean", "sum"]})
pop_item 

pop_item = pop_item.sort_values(("Age","count"), ascending=False)


pop_item[("Price", "mean")] = pop_item[("Price", "mean")].map("${:.2f}".format)
pop_item[("Price", "sum")] = pop_item[("Price", "sum")].map("${:.2f}".format)

pop_item = pop_item.head(5)
pop_item
```
